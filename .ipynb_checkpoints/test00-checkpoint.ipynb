# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def exponential(x,a,b):
    return a*x**b

x = np.linspace(0,5,100)
a = 2
b = 0.5
y = exponential(x,a,b)
plt.plot(x,y)
```

```python
def logarithmic(x,a,b):
    return a*np.log(x*b)
```

```python
z = logarithmic(x,a,b)
plt.plot(x,z)
```
